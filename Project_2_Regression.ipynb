```python
dataset_train_drop =  dataset_train.drop(['yield_anomaly'], axis=1)
dataset_train_drop
```

```python
x = dataset_train_drop.iloc[:,:-1].values
pd.DataFrame(x)
```

```python
y = dataset_train_drop.iloc[:,-1:].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
regressor_lr = LinearRegression()
regressor_lr.fit(x_train,y_train)
```

```python
y_pred = regressor_lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor_dt  = DecisionTreeRegressor()
regressor_dt.fit(x_train,y_train)
```

```python
y_pred = regressor_dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_rfr = RandomForestRegressor(n_estimators=5)
regressor_rfr.fit(x_train,y_train)
```

```python
y_pred = regressor_rfr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
