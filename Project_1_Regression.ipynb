```python
dataset_train_revenue = dataset_train.iloc[:,-1:]
dataset_train_revenue
```

```python
dataset_train =  dataset_train.drop(['date_time'], axis=1)
```

```python
dataset_train =  dataset_train.drop(['traffic_volume','weather_type','weather_description'], axis=1)
dataset_train
```

```python
x = dataset_train.iloc[:,:].values
x
```

```python
from sklearn.preprocessing import  LabelEncoder, OneHotEncoder
labelencoder_x = LabelEncoder()
x[:,0] = labelencoder_x.fit_transform(x[:,0])
#x[:,11] = labelencoder_x.fit_transform(x[:,11])
#x[:,12] = labelencoder_x.fit_transform(x[:,12])
onehotendoer = OneHotEncoder(categorical_features=[0])
#onehotendoer = OneHotEncoder(categorical_features=[11])
#onehotendoer = OneHotEncoder(categorical_features=[12])
x = onehotendoer.fit_transform(x).toarray()
```

```python
y = dataset_train_revenue.iloc[:,:].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3)
```

```python
from sklearn.linear_model import LinearRegression
regressor_lr = LinearRegression()
regressor_lr.fit(x_train,y_train)
```

```python
y_pred = regressor_lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor_dt  = DecisionTreeRegressor()
regressor_dt.fit(x_train,y_train)
```

```python
y_pred = regressor_dt.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor_rfr = RandomForestRegressor(n_estimators=5)
regressor_rfr.fit(x_train,y_train)
```

```python
y_pred = regressor_rfr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```
