```python
import pandas as pd
import numpy as np
import os
```

```python
dataset = pd.read_csv('voice.csv')
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
x = dataset.iloc[:,:-1].values
pd.DataFrame(x)
```

```python
y = dataset.iloc[:,-1:].values
y
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x = sc_x.fit_transform(x)
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
# Classification_Logistic_Regression_Algorithm
```

```python
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train, y_train)
```

```python
y_pred = logmodel.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
# Classification_K-Neighbors_Classifier_Algorithm
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier_knn = KNeighborsClassifier(n_neighbors=5, 
                                      metric='minkowski', p=2)
classifier_knn.fit(x_train, y_train)
```

```python
y_pred = classifier_knn.predict(x_test)
y_pred
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Classification_Naive_Bayes_Algorithm
```

```python
from sklearn.naive_bayes import GaussianNB
classifier_nb = GaussianNB()
classifier_nb.fit(x_train, y_train)
```

```python
y_pred=classifier_nb.predict(x_test)
y_pred
```

```python
confusion_matrix(y_test, y_pred)
```

```python
# Classification_SVC(Support Vector Machine)_Algorithm:-kernel='sigmoid'
```

```python
from sklearn.svm import SVC
classifier_svm = SVC(kernel='sigmoid')
classifier_svm.fit(x_train,y_train)
```

```python
y_pred = classifier_svm.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
y_pred = classifier_svm.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```
